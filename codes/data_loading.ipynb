```python
import pandas as pd
import numpy as np
import h5py
```

```python
directory_path = "/home/draco/work/Thesis/quijote/Quijote_LH/"
```

```python
i_kmax = 80
```

```python
# Load power spectrum
P0 = []
k = []
with h5py.File(directory_path+"LH_equilateral/Pk/Pk_h_RS0_Mmin3.2e13_CIC512_z=0_0-999.h5") as f:
    for i in range (1000):
        k.append(f['k'][:])
        P0.append(f['Pk0/'+str(i)][:i_kmax]) ## read the power spectrum monopole of the i-th map
    #Pk2 = f['Pk2/'+str(i)][:i_kmax] ## read the power spectrum quadrupole of the i-th map
    #Pk4 = f['Pk4/'+str(i)][:i_kmax] ## read the power spectrum quadrupole of the i-th map
P0_df = pd.DataFrame(P0, columns=[f'P0_{i}' for i in range(len(P0[0]))])
k_df = pd.DataFrame(k)



# Load marked power spectrum. For "historical" reasons MP0_3 is just called MP0
MP0 = []
R = 20
p = 1
ds = 0.5
with h5py.File(directory_path+f"LH_equilateral/MPk/MPk_h_RS0_Mmin3.2e13_CIC512_R{R}_p{p}_ds{ds}_z=0_0-999.h5") as mf:
    for i in range (1000):
        MP0.append(mf['MPk0/'+str(i)][:i_kmax]) ## read the power spectrum quadrupole of the i-th map
    #MPk2 = mf['MPk2/'+str(i)][:i_kmax] ## read the power spectrum quadrupole of the i-th map
    #MPk4 = mf['MPk4/'+str(i)][:i_kmax] ## read the power spectrum quadrupole of the i-th map
MP0_df = pd.DataFrame(MP0,columns=[f'MP0_{i}' for i in range(len(MP0[0]))])

MP01 = []
R = 30
p = 1
ds = 0.1
with h5py.File(directory_path+f"LH_equilateral/MPk/MPk_h_RS0_Mmin3.2e13_CIC512_R{R}_p{p}_ds{ds}_z=0_0-999.h5") as mf:
    for i in range (1000):
        MP01.append(mf['MPk0/'+str(i)][:i_kmax]) ## read the power spectrum quadrupole of the i-th map
    #MPk2 = mf['MPk2/'+str(i)][:i_kmax] ## read the power spectrum quadrupole of the i-th map
    #MPk4 = mf['MPk4/'+str(i)][:i_kmax] ## read the power spectrum quadrupole of the i-th map
MP01_df = pd.DataFrame(MP01,columns=[f'MP01_{i}' for i in range(len(MP01[0]))])

MP02 = []
R = 25
p = 1
ds = 0.25
with h5py.File(directory_path+f"LH_equilateral/MPk/MPk_h_RS0_Mmin3.2e13_CIC512_R{R}_p{p}_ds{ds}_z=0_0-999.h5") as mf:
    for i in range (1000):
        MP02.append(mf['MPk0/'+str(i)][:i_kmax]) ## read the power spectrum quadrupole of the i-th map
    #MPk2 = mf['MPk2/'+str(i)][:i_kmax] ## read the power spectrum quadrupole of the i-th map
    #MPk4 = mf['MPk4/'+str(i)][:i_kmax] ## read the power spectrum quadrupole of the i-th map
MP02_df = pd.DataFrame(MP02,columns=[f'MP02_{i}' for i in range(len(MP02[0]))])

MP04 = []
R = 30
p = 1
ds = 0.5
with h5py.File(directory_path+f"LH_equilateral/MPk/MPk_h_RS0_Mmin3.2e13_CIC512_R{R}_p{p}_ds{ds}_z=0_0-999.h5") as mf:
    for i in range (1000):
        MP04.append(mf['MPk0/'+str(i)][:i_kmax]) ## read the power spectrum quadrupole of the i-th map
    #MPk2 = mf['MPk2/'+str(i)][:i_kmax] ## read the power spectrum quadrupole of the i-th map
    #MPk4 = mf['MPk4/'+str(i)][:i_kmax] ## read the power spectrum quadrupole of the i-th map
MP04_df = pd.DataFrame(MP04,columns=[f'MP04_{i}' for i in range(len(MP04[0]))])

# Load bispectrum
B = []
with h5py.File(directory_path+"LH_equilateral/Bisp/Bisp_h_RS0_Mmin3.2e13_PCS256_kmin0.0_kmax0.5_poly10+grav+loc_z=0_0-999.h5") as bf:
    for i in range (1000):
        B.append(np.concatenate((bf["gravity/"+str(i)][:], bf["polynomials/"+str(i)][:]), axis=0)) ## bispectrum of the i-th map
B_df = pd.DataFrame(B,columns=[f'B_{i}' for i in range(len(B[0]))])




# Load marked bispectrum.  For "historical" reasons MB_3 is just called MB

MB = []
R = 20
p = 1
ds = 0.5
with h5py.File(directory_path+f"LH_equilateral/MBisp/MBisp_h_RS0_Mmin3.2e13_PCS256_R{R}_p{p}_ds{ds}_kmin0.0_kmax0.5_poly10_z=0_0-999.h5") as bf:
    for i in range (1000):
        MB.append(bf["polynomials/"+str(i)][:]) ## bispectrum of the i-th map
MB_df = pd.DataFrame(MB,columns=[f'MB_{i}' for i in range(len(MB[0]))])



MB1 = []
R = 30
p = 1
ds = 0.1
with h5py.File(directory_path+f"LH_equilateral/MBisp/MBisp_h_RS0_Mmin3.2e13_PCS256_R{R}_p{p}_ds{ds}_kmin0.0_kmax0.5_poly10_z=0_0-999.h5") as bf:
    for i in range (1000):
        MB1.append(bf["polynomials/"+str(i)][:]) ## bispectrum of the i-th map
MB1_df = pd.DataFrame(MB1,columns=[f'MB1_{i}' for i in range(len(MB1[0]))])



MB2 = []
R = 25
p = 1
ds = 0.25
with h5py.File(directory_path+f"LH_equilateral/MBisp/MBisp_h_RS0_Mmin3.2e13_PCS256_R{R}_p{p}_ds{ds}_kmin0.0_kmax0.5_poly10_z=0_0-999.h5") as bf:
    for i in range (1000):
        MB2.append(bf["polynomials/"+str(i)][:]) ## bispectrum of the i-th map
MB2_df = pd.DataFrame(MB2,columns=[f'MB2_{i}' for i in range(len(MB2[0]))])



MB4 = []
R = 30
p = 1
ds = 0.5
with h5py.File(directory_path+f"LH_equilateral/MBisp/MBisp_h_RS0_Mmin3.2e13_PCS256_R{R}_p{p}_ds{ds}_kmin0.0_kmax0.5_poly10_z=0_0-999.h5") as bf:
    for i in range (1000):
        MB4.append(bf["polynomials/"+str(i)][:]) ## bispectrum of the i-th map
MB4_df = pd.DataFrame(MB4,columns=[f'MB4_{i}' for i in range(len(MB4[0]))])




# Load HMF
HMF = []
with h5py.File(directory_path+"LH_equilateral/HMF/HMF_z=0_0-999.h5") as hf:
    for i in range (1000):
        HMF.append(hf['mass/'+str(i)][:])
HMF = np.array(HMF)
HMF = HMF[:,2:-1]
HMF_df = pd.DataFrame(HMF,columns=[f'HMF_{i}' for i in range(len(HMF[0]))])

```

```python
# parameter labels

label_names = ["fnl_equil", "Om", "h", "ns", "s8"]
params = pd.read_csv(directory_path + "LH_equilateral/latin_hypercube_params.txt", names=label_names, delimiter=r'\s+')
params.to_csv("unnoramlised_lbl.csv",index=False)
#normalising the parameters
mean_params = params.mean(axis=0).values
std_params = params.std(axis = 0).values
norm_label = (params - mean_params)/std_params
```

```python
df = pd.concat([norm_label,P0_df, MP01_df, MP02_df, MP0_df, MP04_df, B_df, MB1_df, MB2_df, MB_df, MB4_df, HMF_df],axis = 1)
```
